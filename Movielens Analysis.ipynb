```python
def genres (a):
    if a['Action'] == 1 :
      return 'Action'
    if a['Adventure'] == 1 :
      return 'Adventure'
    if a['Animation'] == 1 :
      return 'Animation'
    if a['Childrens'] == 1 :
      return 'Childrens'
    if a['Comedy'] == 1 :
      return 'Comedy'
    if a['Crime'] == 1 :
      return 'Crime'
    if a['Documentary'] == 1 :
      return 'Documentary'
    if a['Drama'] == 1 :
      return 'Drama'
    if a['Fantasy'] == 1 :
      return 'Fantasy'
    if a['Film-Noir'] == 1 :
      return 'Film-Noir'
    if a['Horror'] == 1 :
      return 'Horror'
    if a['Musical'] == 1 :
      return 'Musical'
    if a['Mystery'] == 1 :
      return 'Mystery'
    if a['Romance'] == 1 :
      return 'Romance'
    if a['Sci-Fi'] == 1 :
      return 'Sci-Fi'
    if a['Thriller'] == 1 :
      return 'Thriller'
    if a['War'] == 1 :
      return 'War'
    if a['Western'] == 1 :
      return 'Western'
    return 'Other'
```

```python
movies['Genre'] = movies.apply(genres,axis = 1)
```

### 2. Genres Popularity over the years

```python
movielens['releaseyear'] = movielens['releaseyear'].astype('str').astype('int64')
```

```python
movielens.groupby(['releaseyear','Genre']).sum()
```

```python
Genre_Year = pd.crosstab(index=movielens['releaseyear'],columns=movielens['Genre'])
```

```python
Genre_Year
```

```python
Genre_Year.plot(kind="bar",figsize=(15,15),stacked=True)
```

### 3. Top 25 Movies by Average Rating for movies having atleast 100 ratings

```python
movielens.groupby('movietitle')['rating'].size()
```

```python
agg_ratings = movielens.groupby('movietitle').agg({'rating':[np.mean,np.size]})
```

```python
agg_ratings.sort_values([('rating','mean')],ascending = False)
```

```python
top25 = agg_ratings[agg_ratings['rating']['size']>=100].sort_values([('rating','mean')],ascending = False)[:25]
```

```python
#top 25 movies 
top25
```

```python
top25['rating']['mean'].plot(kind="barh",figsize=(15,15))
```
